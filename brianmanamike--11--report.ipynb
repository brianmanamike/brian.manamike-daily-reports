# Exercise 1:

Create a simple line plot for y = x for x values ranging from 0 to 10.




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10)
y = x
# create a figure/canvas of desired size
plt.figure(figsize=(8, 6))

# plot values; with a color `red`
plt.plot(x, y, 'r')

# give labels to the axes
plt.xlabel('x')
plt.ylabel('y')

# give a title to the plot
plt.title(r"Plot of $y=x$")
plt.show()
```

# Exercise 2:

Plot y = x^2 using Matplotlib for x values between -10 and 10.


```python
x = np.linspace(-10, 10)
y = x**2
# create a figure/canvas of desired size
plt.figure(figsize=(8, 6))

# plot values; with a color `red`
plt.plot(x, y, 'r')

# give labels to the axes
plt.xlabel('x')
plt.ylabel('y')

# give a title to the plot
plt.title(r"Plot of $y=x^2$")
plt.show()
```

# Exercise 3:

Create a scatter plot with 50 random points where x and y values are between 0 and 10.


```python
# set figure size and dpi (dots per inch)
plt.figure(figsize=(10, 6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(111)

n = 50
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

# color is given by the angle between X & Y
T = np.arctan2(Y,X)

plt.axes([0.025, 0.025, 0.95, 0.95])

# The alpha blending value, between 0 (transparent) and 1 (opaque).
# s - marker size
# c - color
plt.scatter(X,Y, s=75, c=T, alpha=.5)

plt.xlim(-2.0, 2.0), plt.xticks([])
plt.ylim(-2.0, 2.0), plt.yticks([])
plt.show()
```

# Exercise 4:

Plot a sine wave using Matplotlib for x values between 0 and 2Ï€.


```python
# 256 linearly spaced values between -pi and +pi
# both endpoints would be included; check (X[0], X[-1])
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)

# compute cosine and sin values
C, S = np.cos(X), np.sin(X)

# plot both curves
plt.plot(X, C)
plt.plot(X, S)

# show the plot
plt.show()
```

# Exercise 5:

Create a bar chart for the following data: categories = ['A', 'B', 'C', 'D'] and values = [3, 7, 5, 9].


```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]

# Create bar chart
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='red')

# Add title and labels
plt.title("Bar Chart for Categories A, B, C, and D")
plt.xlabel("Categories")
plt.ylabel("Values")

# Show plot
plt.show()
```

# Exercise 6:

Create a histogram of 1000 random numbers generated from a normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
n = np.random.randn(1000)

# Create a single plot
plt.figure(figsize=(8, 6))

# Plot default histogram
plt.hist(n, color="red")
plt.title("Default Histogram")
plt.xlim((np.min(n), np.max(n)))

# Show plot
plt.tight_layout()
plt.show()

```

# Exercise 7:

Plot two line plots on the same graph: y1 = x^2 and y2 = x^3 for x ranging from 0 to 10.


```python


# Generate values for x
x = np.linspace(0, 10)

# Calculate y for both functions
y_square = x**2
y_cube = x**3

# Create a figure/canvas of desired size
plt.figure(figsize=(8, 6))

# Plot y = x^2 with red color and y = x^3 with blue color
plt.plot(x, y_square, 'r', label=r"$y=x^2$")
plt.plot(x, y_cube, 'b', label=r"$y=x^3$")

# Add labels to the axes
plt.xlabel('x')
plt.ylabel('y')

# Give a title to the plot
plt.title("Plot of $y=x^2$ and $y=x^3$")

# Add a legend to differentiate between the two plots
plt.legend()

# Show the plot
plt.show()
```
